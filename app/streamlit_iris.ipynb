```python
import pandas as pd
from pycaret.regression import *
```

```python
from sklearn.datasets import load_iris
data=load_iris()
```

```python
iris = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
iris["target"] = data.target
```

```python
iris.columns
```

```python
iris.head()
```

```python
regexperiment = setup(iris, target="target")
```

```python
compare_models()
```

```python
model_ada = create_model("ada")
```

```python
tunedgada=tune_model(model_ada)
```

```python
evaluate_model(tunedgada)
```

```python
predictions = predict_model(tunedgada, data=iris)
```

```python
modelo_final = finalize_model(tunedgada)
```

```python
save_model(modelo_final, "model")
```
